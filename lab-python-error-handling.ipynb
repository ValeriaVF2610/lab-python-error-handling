````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Indicate the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print(f"Invalid quantity of {product}s! Please enter a non-negative value.")
            except ValueError:
                print("Invalid input! Please use only numbers to indicate the quantity of products.")
    print("INITIAL INVENTORY\n")
    for product, amount in inventory.items():
        print(f"Product: {product} - Amount: {amount}")
    return inventory
````
